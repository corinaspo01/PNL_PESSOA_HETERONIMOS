```python
import gensim
import string
from nltk.tokenize import WordPunctTokenizer
from urllib import request

# Carga tus datos. (Supongo que tienes una función llamada leer_url que obtiene el texto)

def leer_url(url):
    response = request.urlopen(url)
    text = response.read().decode('utf-8')
    return text

url = "https://raw.githubusercontent.com/corinaspo01/tfm1/main/campos.txt"
texto = leer_url(url)

# Tokeniza tus frases usando NLTK.
excluir = ["el", "como", "cuando", "más", "así", "todas", "tan", "mas", "eso", "le", "qué", "sólo", "porque", "i", "del", "se", "es", "al", "me", "son", "»", "«", "te", "el", "si", "que", "ya", "lo", "ti", "no", "ni", "sí", "la", "los", "las", "un", "una", "unos", "unas", "y", "o", "pero", "por", "para", "con", "sin", "de", "a", "en", "si", "mi", "mis", "tu", "tus", "su", "sus", "¡", "...", "-,", "¿"]

def limpiar_palabras(texto):
    tokenizer = WordPunctTokenizer()
    tokens = tokenizer.tokenize(texto)
    tokens_limpio = [token.lower() for token in tokens if token not in string.punctuation]
    tokens_filtrados = [token for token in tokens_limpio if token not in excluir]
    return tokens_filtrados

# Tokenizar el texto
palabras_limpias = limpiar_palabras(texto)

# Tokeniza cada palabra en frases.
frases_tokenizadas = [palabras_limpias]

# Crea un diccionario Gensim a partir de las frases tokenizadas.
dictionary = gensim.corpora.Dictionary(frases_tokenizadas)

# Crea un corpus Gensim a partir del diccionario y las frases tokenizadas.
corpus = [dictionary.doc2bow(frase) for frase in frases_tokenizadas]

# Crea y entrena un modelo LSI.
num_topics_lsi = 30
lsi_model = gensim.models.LsiModel(corpus, id2word=dictionary, num_topics=num_topics_lsi)

# Imprime los tópicos de LSI.
print("LSI Topics:")
lsi_topics = lsi_model.print_topics(num_words=30)
for rtopic in lsi_topics:
    print(rtopic)
```

```python
# Crea y entrena un modelo LSI.
num_topics_lsi = 30
lsi_model = gensim.models.LsiModel(corpus, id2word=dictionary, num_topics=num_topics_lsi)

# Obtén las palabras clave de los tópicos de LSI.
lsi_topics = lsi_model.show_topics(num_topics=num_topics_lsi, num_words=30)

# Concatena todas las palabras clave de los tópicos.
all_keywords = " ".join([topic[1] for topic in lsi_topics])

# Crea un objeto WordCloud.
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(all_keywords)

# Muestra el WordCloud.
plt.figure(figsize=(5, 5), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

# Guarda la imagen del WordCloud.
plt.savefig("lsi_wordcloud.png")
plt.show()
```
